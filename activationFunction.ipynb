```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Define the activation functions 
def linear(x): 
    return x 

#sigmoid 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

#thanh(x)
def tanh(x):
    return np.tanh(x)

def relu(x): 
    return np.maximum(0, x) 

def softmax(x): 
    e_x = np.exp(x - np.max(x)) 
    return e_x / e_x.sum(axis=0, keepdims=True) 

```

```python
#Generate data 
x = np.linspace(-10, 10, 400) 
```

```python
# Plotting the graphs for each activation function 
fig, axs = plt.subplots(3, 2, figsize=(12, 12)) 

 
# Linear Activation 
axs[0, 0].plot(x, linear(x)) 
axs[0, 0].set_title("Linear Activation: f(x) = x") 
axs[0, 0].set_xlabel("x") 
axs[0, 0].set_ylabel("f(x)") 

#Sigmoid Activation 
axs[0, 1].plot(x, sigmoid(x)) 
axs[0, 1].set_title("Sigmoid Activation: f(x) = 1 / (1 + e^(-x))") 
axs[0, 1].set_xlabel("x") 
axs[0, 1].set_ylabel("f(x)") 

# Tanh Activation 
axs[1, 0].plot(x, tanh(x)) 
axs[1, 0].set_title("Tanh Activation: f(x) = tanh(x)") 
axs[1, 0].set_xlabel("x") 
axs[1, 0].set_ylabel("f(x)") 

# ReLU Activation 
axs[1, 1].plot(x, relu(x)) 
axs[1, 1].set_title("ReLU Activation: f(x) = max(0, x)") 
axs[1, 1].set_xlabel("x") 
axs[1, 1].set_ylabel("f(x)") 

# Softmax Activation (for a vector of values) 
x_softmax = np.linspace(-2, 2, 5) 
softmax_values = softmax(x_softmax) 
axs[2, 0].bar(x_softmax, softmax_values, color="orange") 
axs[2, 0].set_title("Softmax Activation: f(x) = exp(x) / sum(exp(x))") 
axs[2, 0].set_xlabel("x") 
axs[2, 0].set_ylabel("Probability") 

#Remove the empty subplot 
axs[2, 1].axis("off") 

# Adjust layout 
plt.tight_layout() 
plt.show() 
```
