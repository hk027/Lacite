### comparaison des fonctions de perte de régression (MSE, MAE, Huber) en fonction de l’erreur.

```python
import numpy as np
import matplotlib.pyplot as plt

# Erreur
e = np.linspace(-5, 5, 400)
print(e)
# Fonctions de perte
mse = e**2
mae = np.abs(e)

delta = 1.0
huber = np.where(np.abs(e) <= delta,
                 0.5 * e**2,
                 delta * (np.abs(e) - 0.5 * delta))

# Tracé
plt.figure(figsize=(8,5))
plt.plot(e, mse, label="MSE", linewidth=2)
plt.plot(e, mae, label="MAE", linewidth=2)
plt.plot(e, huber, label="Huber loss", linewidth=2)

plt.xlabel("Erreur")
plt.ylabel("Perte")
plt.title("Comparaison des fonctions de perte (régression)")
plt.legend()
plt.grid(True)

plt.show()
```
